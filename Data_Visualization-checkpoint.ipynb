## Let's Apply K-Means Clustering 
### We are using K-Means Clustering to understand our data and not to predict anything

### Finding the optimal number of clusters using the elbow method. 
It executes the K-means clustering on a given dataset for different K values (ranges from 1-10).
For each value of K, calculates the WCSS (Within Cluster Sum of Squares) value.
Plots a curve between calculated WCSS values and the number of clusters K.
The sharp point of bend or a point of the plot looks like an arm, then that point is considered as the best value of K.

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(final_data)
    wcss.append(kmeans.inertia_)
```

```python
figure(figsize=(8,4), dpi=100)
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

### From the above plot, we can see the elbow point is at 3. So the number of clusters here will be 3.

### Training model to visulize our form clusters

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(final_data)
```

### Visualising the clusters

```python
x = np.array(x)
figure(figsize=(20,10), dpi=300)
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'cyan')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.title('Clustering')  
plt.legend()  
plt.show()  
```

### The output image is clearly showing the three different clusters with different colors. Where yellow is a centriod.
